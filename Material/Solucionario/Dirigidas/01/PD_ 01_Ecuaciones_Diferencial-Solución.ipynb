##**Ejercicio 2.1 a**: Resolver la siguiente ecuación diferencial
$\frac{dy}{dt}+4y=-20\qquad y(0)=10$

```python
#========================================================================
# PARTE 1: Definir la ecuación diferencial dt/dy
#========================================================================

### 1.1 Ecuación
def model(y,t):
    k = 4
    dydt = -20-k*y
    return dydt

#========================================================================
# PARTE 1: Definiendo valores
#========================================================================

### 2.1 Establcer las condiciones iniciales
y0 = [10]

### 2.2 Instantes del tiempo
t = np.linspace(0,10,200)

### 2.3 Resolver la ecuación diferencial
y = odeint(model,y0,t)

### 2.4 Mostramos los 6 valores principales
for i in range(0,7):
  print(f'y({t[i]}) = {y[i,0]}')

#========================================================================
# PARTE 3: Graficar resultados
#========================================================================

### 3.1 Labels
plt.title("Convergencia", fontsize=20)
plt.xlabel('time', fontsize=20)
plt.ylabel('y(t)', fontsize=20)

### 3.2 Cuadrícula
plt.grid()

### 3.3 Tamaño
plt.rcParams["figure.figsize"] = (20, 15)

### 3.4 Línea de convergencia
plt.axhline(y=-5, color='r', linestyle='-')

### 3.6 Ajusate de ejes
plt.ylim(-6,11)
plt.xlim(right=5)

### 3.7 Curva
plt.plot(t,y)

### 3.8 Mostrar el gráfico
plt.show()
```

##**Ejercicio 2.1 c**: Resolver la siguiente ecuación diferencial
$\frac{dy}{dt}+3y=6t; \qquad y(0)=\frac{1}{2}$

```python
#===============================================================================
# PARTE 1: Definir la ecuación diferencial dt/dy
#===============================================================================
mt = np.array([0,50])

sample_times = np.arange(len(mt))
tfunc = interp1d(sample_times, mt, bounds_error=False, fill_value="extrapolate")

# Test ODE function
def test(y, t):
    dydt = 6*tfunc(t)-3*y
    return dydt

#===============================================================================
# PARTE 2: Definiendo valores
#===============================================================================

### 2.1 Establcer las condiciones iniciales
y0 = [1/3]

### 2.2 Instantes del tiempo
tspan = np.linspace(0,2,200)

### 2.3 Resolver la ecuación diferencial
yt = odeint(test, y0, tspan)

### 2.4 Mostramos los 6 valores principales
for i in range(0,7):
  print(f'y({t[i]}) = {y[i,0]}')

#===============================================================================
# PARTE 3: Graficar resultados
#===============================================================================

### 3.1 Labels
plt.title("Divergencia", fontsize=20)
plt.xlabel('time', fontsize=20)
plt.ylabel('y(t)', fontsize=20)

### 3.2 Cuadrícula
plt.grid()

### 3.3 Tamaño

### 3.4 Línea de convergencia
plt.axhline(y=1/3, color='r', linestyle='-')

###3.5 Ajuste de ejes
plt.ylim(top=3)
plt.xlim(0,0.2)

### 3.6 Curva
plt.plot(tspan, yt, 'black')

### 3.7 Mostrar el gráfico
plt.show()
```

##**Ejercicio 2.2 a**: Resolver la siguiente ecuación diferencial
$y''(t)+y'(t)+\frac{1}{4}y(t)=9;\quad y(0)=30\quad$ y $\quad y'(0)=15$

```python
#========================================================================
# PARTE 1: Definir la ecuación diferencial dt/dy
#========================================================================

### 1.1 Ecuación
def f(u,t):
  return (u[1],-u[1]-0.25*u[0]+9)

#========================================================================
# PARTE 2: Definiendo valores
#========================================================================

### 2.1 Establcer las condiciones iniciales
y0=[30,15]

### 2.2 Instantes del tiempo
ts=np.linspace(0,40,100)

### 2.3 Resolver la ecuación diferencial
us=odeint(f,y0,ts)
ys=us[:,0]

### 2.4 Mostramos los 6 valores principales
for i in range(0,31):
  print(f'y({i}) = {ys[i]}')

#========================================================================
# PARTE 3: Graficar resultados
#========================================================================

### 3.1 Labels
plt.title("Convergencia", fontsize=20)
plt.xlabel('time', fontsize=20)
plt.ylabel('y(t)', fontsize=20)

### 3.2 Cuadrícula
plt.grid()

### 3.3 Tamaño
plt.rcParams["figure.figsize"] = (10, 5)

### 3.4 Línea de convergencia
plt.axhline(y=36, color='r', linestyle='-')

### 3.5 Curva
plt.plot(ts,ys,'-')
### 3.6 Mostrar el gráfico
plt.show()
```
